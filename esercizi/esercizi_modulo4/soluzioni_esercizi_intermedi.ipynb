## Esercizio 1.1: Conversione Sicura

**Obiettivo:** Creare una funzione che converte una stringa in numero gestendo gli errori.

```python
# Soluzione Esercizio 1.1

def converti_numero(testo):
    """
    Converte una stringa in numero (int o float).
    Restituisce None se la conversione fallisce.
    """
    # Prima prova a convertire in int
    try:
        return int(testo)
    except ValueError:
        pass
    
    # Se fallisce, prova float
    try:
        return float(testo)
    except ValueError:
        return None

# Test
print(converti_numero("42"))      # 42 (int)
print(converti_numero("3.14"))    # 3.14 (float)
print(converti_numero("ciao"))    # None
print(converti_numero("-10"))     # -10 (int)
print(converti_numero("1e5"))     # 100000.0 (float)
```

**Spiegazione:**
- Proviamo prima `int()` perchÃ© Ã¨ piÃ¹ restrittivo (non accetta decimali)
- Se `int()` fallisce con `ValueError`, proviamo `float()`
- Se anche `float()` fallisce, restituiamo `None` invece di far crashare il programma

## Esercizio 1.2: Accesso Sicuro a Dizionario

**Obiettivo:** Gestire KeyError quando si accede a chiavi inesistenti.

```python
# Soluzione Esercizio 1.2

utenti = {
    "mario": {"eta": 30, "email": "mario@email.it"},
    "luigi": {"eta": 25}
}

def get_email_utente(username):
    """
    Restituisce l'email dell'utente.
    Gestisce: utente non esistente, email non presente.
    """
    try:
        utente = utenti[username]
    except KeyError:
        return "Utente non trovato"
    
    try:
        return utente["email"]
    except KeyError:
        return "Email non disponibile"

# Test
print(get_email_utente("mario"))   # mario@email.it
print(get_email_utente("luigi"))   # Email non disponibile
print(get_email_utente("peach"))   # Utente non trovato
```

**Spiegazione:**
- La versione con try-except annidati separa chiaramente i due tipi di errore
- In alternativa, si puÃ² usare `.get()`: `utenti.get(username, {}).get("email")`

## Esercizio 1.3: Validatore con Eccezione Custom

**Obiettivo:** Creare un'eccezione personalizzata per validare password.

```python
# Soluzione Esercizio 1.3

class PasswordDebolError(Exception):
    """Eccezione per password non sicure."""
    pass

def valida_password(password):
    """
    Valida una password. Solleva PasswordDebolError se:
    - Meno di 8 caratteri
    - Nessun numero
    - Nessuna maiuscola
    """
    if len(password) < 8:
        raise PasswordDebolError("La password deve avere almeno 8 caratteri")
    
    if not any(c.isdigit() for c in password):
        raise PasswordDebolError("La password deve contenere almeno un numero")
    
    if not any(c.isupper() for c in password):
        raise PasswordDebolError("La password deve contenere almeno una maiuscola")
    
    return True

# Test
passwords_test = ["abc123", "Password", "abcdefgh", "Password123"]

for pwd in passwords_test:
    try:
        valida_password(pwd)
        print(f"'{pwd}' â†’ âœ… Valida")
    except PasswordDebolError as e:
        print(f"'{pwd}' â†’ âŒ {e}")
```

## Esercizio 2.1: Generatore di Password

**Obiettivo:** Usare i moduli `random` e `string` per generare password.

```python
# Soluzione Esercizio 2.1

import random
import string

def genera_password(lunghezza=12, con_speciali=True):
    """Genera una password casuale."""
    caratteri = string.ascii_letters + string.digits
    if con_speciali:
        caratteri += string.punctuation
    
    password = ''.join(random.choice(caratteri) for _ in range(lunghezza))
    return password

# Test
print("Password senza speciali (8 car.):")
for _ in range(3):
    print(f"  {genera_password(8, False)}")

print("\nPassword con speciali (16 car.):")
for _ in range(3):
    print(f"  {genera_password(16, True)}")
```

## Esercizio 2.3: Salvataggio Configurazione

**Obiettivo:** Usare il modulo `json` per salvare e caricare configurazioni.

```python
# Soluzione Esercizio 2.3

import json
import os

def carica_config(file_path):
    """
    Carica configurazione da file JSON.
    Restituisce dizionario vuoto se il file non esiste.
    """
    try:
        with open(file_path, "r") as f:
            return json.load(f)
    except FileNotFoundError:
        return {}
    except json.JSONDecodeError:
        print("Attenzione: file corrotto")
        return {}

def salva_config(config, file_path):
    """Salva configurazione su file JSON."""
    with open(file_path, "w") as f:
        json.dump(config, f, indent=2)

def aggiorna_config(file_path, chiave, valore):
    """Carica config, aggiorna una chiave, e salva."""
    config = carica_config(file_path)
    config[chiave] = valore
    salva_config(config, file_path)
    return config

# Test
FILE_CONFIG = "test_config.json"

salva_config({"tema": "dark", "lingua": "it"}, FILE_CONFIG)
config = carica_config(FILE_CONFIG)
print(f"Config: {config}")
print(f"Tema: {config['tema']}")

aggiorna_config(FILE_CONFIG, "volume", 75)
print(f"Dopo update: {carica_config(FILE_CONFIG)}")

os.remove(FILE_CONFIG)
```

**Spiegazione:**
- `carica_config()` usa try-except per gestire file mancanti o corrotti
- `salva_config()` scrive il dizionario su file JSON
- `aggiorna_config()` combina le due operazioni

## Esercizio 3.1: Classe Prodotto

**Obiettivo:** Creare una classe con metodi speciali.

```python
# Soluzione Esercizio 3.1

class Prodotto:
    def __init__(self, nome, prezzo, quantita=1):
        self.nome = nome
        self.prezzo = prezzo
        self.quantita = quantita
    
    def __str__(self):
        return f"{self.nome}: {self.prezzo}â‚¬ x {self.quantita}"
    
    def __repr__(self):
        return f"Prodotto('{self.nome}', {self.prezzo}, {self.quantita})"
    
    def __eq__(self, altro):
        if not isinstance(altro, Prodotto):
            return False
        return self.nome == altro.nome
    
    def valore_totale(self):
        return self.prezzo * self.quantita

# Test
p1 = Prodotto("Laptop", 999, 2)
p2 = Prodotto("Laptop", 1200, 1)
p3 = Prodotto("Mouse", 25, 5)

print(p1)
print(p1.valore_totale())
print(f"p1 == p2: {p1 == p2}")
print(f"p1 == p3: {p1 == p3}")
```

## Esercizio 3.2: Gerarchia Forme

**Obiettivo:** Usare ereditarietÃ  e polimorfismo.

```python
# Soluzione Esercizio 3.2

import math

class Forma:
    def area(self):
        raise NotImplementedError
    
    def perimetro(self):
        raise NotImplementedError

class Quadrato(Forma):
    def __init__(self, lato):
        self.lato = lato
    
    def area(self):
        return self.lato ** 2
    
    def perimetro(self):
        return 4 * self.lato

class Cerchio(Forma):
    def __init__(self, raggio):
        self.raggio = raggio
    
    def area(self):
        return math.pi * self.raggio ** 2
    
    def perimetro(self):
        return 2 * math.pi * self.raggio

# Test polimorfismo
forme = [Quadrato(5), Cerchio(3)]
for f in forme:
    print(f"Area: {f.area():.2f}, Perimetro: {f.perimetro():.2f}")
```

## Esercizio 3.3: Sistema di Utenti

**Obiettivo:** Usare property, ereditarietÃ  e composizione.

```python
# Soluzione Esercizio 3.3

from datetime import date

class Utente:
    def __init__(self, username, email):
        self.username = username
        self._email = email
        self.data_registrazione = date.today()
    
    @property
    def email(self):
        return self._email
    
    @email.setter
    def email(self, valore):
        if "@" not in valore:
            raise ValueError("Email non valida")
        self._email = valore

class UtenteAdmin(Utente):
    def __init__(self, username, email, livello=1):
        super().__init__(username, email)
        self.livello = livello
    
    def promuovi(self):
        self.livello += 1

# Test
admin = UtenteAdmin("admin", "admin@site.com", livello=1)
print(f"Livello iniziale: {admin.livello}")
admin.promuovi()
admin.promuovi()
print(f"Livello finale: {admin.livello}")
```
