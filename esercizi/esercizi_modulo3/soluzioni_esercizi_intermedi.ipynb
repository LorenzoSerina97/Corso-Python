# ðŸŽ¯ Sezione 1: Controllo di Flusso

## Esercizio 1.1: Verifica Triangolo

**Obiettivo:** Verificare se tre lati formano un triangolo valido e classificarlo.

````python
# Soluzione Esercizio 1.3

lato1, lato2, lato3 = 3, 4, 5

# Regola del triangolo: la somma di due lati deve essere > del terzo
if (lato1 + lato2 > lato3 and 
    lato1 + lato3 > lato2 and 
    lato2 + lato3 > lato1):
    
    print("âœ… I lati formano un triangolo valido!")
    
    # Classificazione del triangolo
    if lato1 == lato2 == lato3:
        tipo = "Equilatero"
    elif lato1 == lato2 or lato2 == lato3 or lato1 == lato3:
        tipo = "Isoscele"
    else:
        tipo = "Scaleno"
    
    print(f"Tipo: {tipo}")
else:
    print("âŒ I lati NON formano un triangolo valido")
````

````python
# Versione funzione con test multipli
def analizza_triangolo(a, b, c):
    """Verifica e classifica un triangolo."""
    # Verifica validitÃ 
    if not (a + b > c and a + c > b and b + c > a):
        return "Non valido"
    
    # Classificazione
    if a == b == c:
        return "Equilatero"
    elif a == b or b == c or a == c:
        return "Isoscele"
    else:
        return "Scaleno"

# Test casi diversi
test_triangoli = [
    (3, 4, 5),    # Scaleno (anche rettangolo!)
    (5, 5, 5),    # Equilatero
    (3, 3, 4),    # Isoscele
    (1, 2, 10),   # Non valido
]

for lati in test_triangoli:
    risultato = analizza_triangolo(*lati)
    print(f"Lati {lati} â†’ {risultato}")
````

## Esercizio 2.1: Controllo Range

**Obiettivo:** Verificare se un voto Ã¨ nell'intervallo valido e determinare la promozione.

````python
# Soluzione Esercizio 2.2

voto = 75

# Modo pythonico: concatenamento confronti
if 0 <= voto <= 100:
    print(f"Voto {voto} Ã¨ valido")
    
    if voto >= 60:
        print("ðŸŽ‰ Promosso!")
    else:
        print("ðŸ“š Bocciato")
else:
    print(f"Voto {voto} non valido (deve essere 0-100)")
````

````python
# Test con vari voti
voti_test = [-5, 45, 60, 85, 105]

for voto in voti_test:
    if 0 <= voto <= 100:
        esito = "Promosso" if voto >= 60 else "Bocciato"
        print(f"Voto {voto:3}: {esito}")
    else:
        print(f"Voto {voto:3}: Non valido")
````

**Spiegazione:**
- `0 <= voto <= 100` Ã¨ equivalente a `voto >= 0 and voto <= 100` ma piÃ¹ leggibile
- Questo Ã¨ un pattern tipico di Python chiamato "chained comparison"

## Esercizio 2.2: Valori Truthy/Falsy

**Obiettivo:** Prevedere l'output senza eseguire il codice.

````python
# Soluzione Esercizio 2.3

lista = []          # Falsy (lista vuota)
nome = "Mario"      # Truthy (stringa non vuota)
numero = 0          # Falsy (zero)

if lista:
    print("A")      # NON eseguito - lista Ã¨ Falsy
if nome:
    print("B")      # âœ… ESEGUITO - nome Ã¨ Truthy
if numero:
    print("C")      # NON eseguito - 0 Ã¨ Falsy
if nome and not lista:
    print("D")      # âœ… ESEGUITO - True and True = True
````

**Risposta:** Vengono stampate le lettere **B** e **D**

**Spiegazione dettagliata:**
- `lista = []` â†’ Falsy (collezione vuota)
- `nome = "Mario"` â†’ Truthy (stringa con contenuto)
- `numero = 0` â†’ Falsy (zero Ã¨ sempre falsy)
- `nome and not lista` â†’ `True and not False` â†’ `True and True` â†’ `True`

````python
# Tabella riassuntiva valori Truthy/Falsy
valori = [
    ([], "lista vuota"),
    ([1, 2], "lista con elementi"),
    ("", "stringa vuota"),
    ("ciao", "stringa con testo"),
    (0, "zero"),
    (42, "numero diverso da zero"),
    (None, "None"),
    ({}, "dizionario vuoto"),
]

print("Valore               | Truthy/Falsy")
print("-" * 40)
for valore, descrizione in valori:
    risultato = "Truthy" if valore else "Falsy"
    print(f"{descrizione:20} | {risultato}")
````

## Esercizio 3.1: Conta le Vocali

**Obiettivo:** Contare le vocali in una stringa.

````python
# Soluzione Esercizio 3.1

frase = "Python e fantastico"
vocali = "aeiouAEIOU"
conteggio = 0

for carattere in frase:
    if carattere in vocali:
        conteggio += 1

print(f"Frase: '{frase}'")
print(f"Vocali trovate: {conteggio}")
````

````python
# Versione con list comprehension (piÃ¹ pythonico)
frase = "Python e fantastico"
vocali = "aeiouAEIOU"

conteggio = sum(1 for c in frase if c in vocali)
print(f"Vocali trovate: {conteggio}")

# Oppure contando i caratteri filtrati
vocali_trovate = [c for c in frase if c in vocali]
print(f"Vocali: {vocali_trovate}")
print(f"Conteggio: {len(vocali_trovate)}")
````

**Spiegazione:**
- `for carattere in frase` itera su ogni singolo carattere della stringa
- `in` verifica se il carattere Ã¨ presente nella stringa delle vocali
- La versione con comprehension Ã¨ piÃ¹ concisa e "pythonica"

## Esercizio 3.2: Numeri Primi

**Obiettivo:** Trovare tutti i numeri primi da 2 a 50.

````python
# Soluzione Esercizio 3.2

print("Numeri primi da 2 a 50:")

for numero in range(2, 51):
    # Verifica se Ã¨ primo controllando i divisori
    for divisore in range(2, numero):
        if numero % divisore == 0:
            # Trovato un divisore: non Ã¨ primo
            break
    else:
        # Il ciclo Ã¨ terminato senza break: Ã¨ primo!
        print(numero, end=" ")
````

````python
# Versione ottimizzata con funzione
def is_primo(n):
    """Verifica se n Ã¨ un numero primo."""
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    # Basta controllare fino alla radice quadrata
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

# Lista dei primi
primi = [n for n in range(2, 51) if is_primo(n)]
print(f"Primi trovati: {primi}")
print(f"Totale: {len(primi)} numeri primi")
````

**Spiegazione:**
- Il costrutto `for...else` Ã¨ una caratteristica poco nota di Python
- L'`else` viene eseguito SOLO se il ciclo termina naturalmente (senza `break`)
- Questo pattern Ã¨ perfetto per cercare elementi con condizioni di uscita anticipata

## Esercizio 3.3: List Comprehension

**Obiettivo:** Convertire cicli tradizionali in list comprehension.

````python
# Soluzione Esercizio 3.3 - Parte 1

# Versione originale con ciclo
parole = ["python", "java", "perl", "ruby", "php"]
risultato = []
for p in parole:
    if p.startswith("p"):
        risultato.append(p)
print(f"Ciclo for: {risultato}")

# Versione list comprehension
risultato = [p for p in parole if p.startswith("p")]
print(f"Comprehension: {risultato}")
````

````python
# Soluzione Esercizio 3.3 - Parte 2

# Versione originale con ciclo
quadrati_pari = []
for n in range(1, 11):
    if n % 2 == 0:
        quadrati_pari.append(n ** 2)
print(f"Ciclo for: {quadrati_pari}")

# Versione list comprehension
quadrati_pari = [n ** 2 for n in range(1, 11) if n % 2 == 0]
print(f"Comprehension: {quadrati_pari}")
````

**Struttura della List Comprehension:**
```python
[espressione for elemento in iterabile if condizione]
```

- `espressione`: cosa mettere nella lista (es: `n ** 2`)
- `elemento`: variabile che assume ogni valore
- `iterabile`: la sequenza da cui prendere i valori
- `condizione` (opzionale): filtro per includere solo certi elementi

## Esercizio 4.2: Lambda e Ordinamento

**Obiettivo:** Usare funzioni lambda per ordinare liste di dizionari.

````python
# Soluzione Esercizio 4.3

prodotti = [
    {"nome": "Laptop", "prezzo": 999, "disponibilita": 5},
    {"nome": "Mouse", "prezzo": 25, "disponibilita": 50},
    {"nome": "Tastiera", "prezzo": 75, "disponibilita": 20},
]

# 1. Ordina per prezzo crescente
per_prezzo = sorted(prodotti, key=lambda x: x["prezzo"])
print("Per prezzo crescente:")
for p in per_prezzo:
    print(f"  {p['nome']}: {p['prezzo']}â‚¬")
````

````python
# 2. Ordina per disponibilitÃ  decrescente
per_disponibilita = sorted(prodotti, key=lambda x: x["disponibilita"], reverse=True)
print("Per disponibilitÃ  decrescente:")
for p in per_disponibilita:
    print(f"  {p['nome']}: {p['disponibilita']} pezzi")
````

````python
# 3. Ordina per nome (alfabetico)
per_nome = sorted(prodotti, key=lambda x: x["nome"])
print("Per nome (alfabetico):")
for p in per_nome:
    print(f"  {p['nome']}")
````

**Spiegazione:**
- `key=lambda x: x["prezzo"]` estrae il valore su cui ordinare
- `reverse=True` inverte l'ordine (decrescente)
- Le lambda sono perfette per operazioni semplici come questa

# ðŸŽ¯ Sezione 5: Scope

## Esercizio 5.1: Prevedi l'Output

**Obiettivo:** Capire come funziona lo scope locale e globale.

````python
# Soluzione Esercizio 5.1

x = 10  # Variabile globale

def funzione_a():
    x = 20  # Variabile LOCALE (shadowing)
    print(f"A: {x}")

def funzione_b():
    print(f"B: {x}")  # Legge la globale

def funzione_c():
    global x  # Dichiara che userÃ  la globale
    x = 30    # Modifica la globale
    print(f"C: {x}")

funzione_a()  # Stampa: A: 20
funzione_b()  # Stampa: B: 10
funzione_c()  # Stampa: C: 30
print(f"Finale: {x}")  # Stampa: Finale: 30
````

**Risposta:**
```
A: 20
B: 10
C: 30
Finale: 30
```

**Spiegazione:**
- `funzione_a`: crea una variabile locale `x = 20`, la globale rimane `10`
- `funzione_b`: legge la variabile globale `x = 10`
- `funzione_c`: usa `global x` per modificare la variabile globale a `30`
- Alla fine, la globale vale `30` perchÃ© Ã¨ stata modificata da `funzione_c`

## Esercizio 5.2: Contatore con Closure

**Obiettivo:** Creare una factory di contatori usando closure e nonlocal.

````python
# Soluzione Esercizio 5.2

def crea_contatore(inizio=0):
    """
    Restituisce una funzione che ad ogni chiamata
    restituisce il prossimo numero della sequenza.
    """
    conteggio = inizio
    
    def conta():
        nonlocal conteggio  # Usa la variabile della funzione esterna
        conteggio += 1
        return conteggio
    
    return conta

# Test
contatore = crea_contatore()
print(contatore())  # 1
print(contatore())  # 2
print(contatore())  # 3

print("---")

altro_contatore = crea_contatore(10)
print(altro_contatore())  # 11
print(altro_contatore())  # 12
````

**Spiegazione:**
- La funzione interna `conta()` "ricorda" la variabile `conteggio` anche dopo che `crea_contatore()` Ã¨ terminata
- `nonlocal` permette di modificare la variabile della funzione esterna
- Ogni chiamata a `crea_contatore()` crea una nuova closure con il proprio stato

## Esercizio 5.3: Correggere il Bug

**Obiettivo:** Identificare e correggere un bug legato allo scope.

````python
# Codice con BUG (non eseguire, genera errore)

# totale_vendite = 0

# def aggiungi_vendita(importo):
#     totale_vendite += importo  # âŒ UnboundLocalError!
#     print(f"Vendita registrata: {importo}â‚¬")

# aggiungi_vendita(100)
````

````python
# SOLUZIONE 1: Usare global (semplice ma sconsigliato)

totale_vendite = 0

def aggiungi_vendita(importo):
    global totale_vendite
    totale_vendite += importo
    print(f"Vendita registrata: {importo}â‚¬")

def mostra_totale():
    print(f"Totale vendite: {totale_vendite}â‚¬")

aggiungi_vendita(100)
aggiungi_vendita(50)
mostra_totale()
````

````python
# SOLUZIONE 2: Passare e restituire il valore (consigliato)

def aggiungi_vendita(totale, importo):
    totale += importo
    print(f"Vendita registrata: {importo}â‚¬")
    return totale

def mostra_totale(totale):
    print(f"Totale vendite: {totale}â‚¬")

# Uso
totale = 0
totale = aggiungi_vendita(totale, 100)
totale = aggiungi_vendita(totale, 50)
mostra_totale(totale)
````

**Spiegazione del bug:**
- `totale_vendite += importo` equivale a `totale_vendite = totale_vendite + importo`
- Python vede un'assegnazione e crea una variabile locale
- Ma la variabile locale non esiste ancora quando cerca di leggerla â†’ `UnboundLocalError`

**Best practice:**
- Evitare `global` quando possibile
- Preferire passare valori come parametri e restituirli
- Per stato complesso, usare classi
